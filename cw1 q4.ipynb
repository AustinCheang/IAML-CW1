### Load the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

### Load the Dataset

```python
dataset = pd.read_csv('./data/classification_eval_1.csv')
dataset_binary = dataset.copy()
y = dataset.iloc[:,0].values
```

#### Cleaning the data with corresponding class with the threshold >= 0.5

```python
# Get all the alg_1 rows which class gt as 1 with the threshold
for i in range(1, dataset_binary.shape[1]):
    dataset_binary.iloc[:,i].mask(dataset.iloc[:,i] >= 0.5, 1, inplace = True)
    dataset_binary.iloc[:,i].mask(dataset.iloc[:,i] < 0.5, 0, inplace = True)
dataset_binary
```

```python
print('alg_1 label = 1: ', len([x for x  in dataset_binary['alg_1'].values if x == 1]))
print('alg_1 label = 0: ', len([x for x  in dataset_binary['alg_1'].values if x == 0]))
print()
print('alg_2 label = 1: ', len([x for x  in dataset_binary['alg_2'].values if x == 1]))
print('alg_2 label = 0: ', len([x for x  in dataset_binary['alg_2'].values if x == 0]))
print()
print('alg_3 label = 1: ', len([x for x  in dataset_binary['alg_3'].values if x == 1]))
print('alg_3 label = 0: ', len([x for x  in dataset_binary['alg_3'].values if x == 0]))
print()
print('alg_4 label = 1: ', len([x for x  in dataset_binary['alg_4'].values if x == 1]))
print('alg_4 label = 0: ', len([x for x  in dataset_binary['alg_4'].values if x == 0]))
```

#### Find the Accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score

for i in range(1, dataset_binary.shape[1]):
    print('alg_'+str(i))
    cm = confusion_matrix(dataset_binary.iloc[:,0], dataset_binary.iloc[:,i])
    print('Confusion Matrix: ')
    print(cm)
    print('Accuracy: {0:.2f}%'.format(accuracy_score(dataset_binary.iloc[:,0], dataset_binary.iloc[:,i])*100))
    print()
```
